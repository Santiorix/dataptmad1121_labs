```python
food_list = ['Bananas', 'Chocolate', 'bread', 'diapers', 'Ice Cream', 'Brownie Mix', 'broccoli']

# Your code here:
food_list_b = [] #Creo la lista vacía para añadir los que empiezan en "b"
for i in food_list: #creo el loop para recorrer la lista
    if i.startswith("b") or i.startswith("B"): #con este condicional selecciono los que empiezan por b o B
        food_list_b.append(i)
#print(food_list_b) #Compropbación
food_list_b.insert(0,"Grocery list:") #añado la palabra "Grocery list:" a la primera posición de la lista o posicimón 0
#print (food_list_b) #comprobación

grocery_list = ", ".join(food_list_b) #creo el string grocery list a partir del listado
#print (grocery_list)
grocery_list_2 = re.sub(":,", ":", grocery_list) #cambio la :, que me ha quedado después de grocery por :
grocery_list_3 = grocery_list_2 + "." #añado el punto final al string
print (grocery_list_3.capitalize()) # resultado, poniendo todo en minúsculas salvo la primera con la función capitalize
```
